```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("uciml/iris")

print("Path to dataset files:", path)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(path + "/Iris.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x=df.index, y="SepalLengthCm", hue="Species", marker="o")
plt.title("Line Plot of Sepal Length")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True)
```

```python
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Sepal Length vs Sepal Width')
plt.show()

```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Species', data=df, palette='Set2')

# Adding labels and title
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Count of Each Species")

# Show the plot
plt.show()
```

```python
sns.boxplot(data=df, x='Species', y='SepalLengthCm', color = "orange")
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```python
numeric_columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Plot histograms for each numeric column
for column in numeric_columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=column, kde=True)
    plt.title(f'Histogram of {column}')
    plt.show()
```

```python
df.head(15).plot(kind="bar", y=["SepalLengthCm", "SepalWidthCm"], figsize=(12, 6), color=["orange", "blue"])

# Labels and title
plt.xlabel("Index(first 15)")
plt.ylabel("Length (cm)")
plt.title("Comparison of Sepal Length and Sepal Width")
plt.legend(["Sepal Length", "Sepal Width"])
plt.show()
```
